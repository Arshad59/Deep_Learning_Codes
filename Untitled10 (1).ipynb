```python
import numpy as np
import matplotlib.pyplot as plt

# Activation Function
def sigmoid(w_sum):
    return 1 / (1 + np.exp(-w_sum))

# Get Prediction
def predict(features, weights, bias):
    return sigmoid(np.dot(features, weights) + bias)

# Loss Function
def cross_entropy(target, pred):
    return -(target * np.log(pred) + (1 - target) * np.log(1 - pred))

# Update Weights
def gradient_descent(features, target, weights, bias, learnrate, pred):
    # Calculate the error
    error = target - pred

    # Update weights
    weights += learnrate * error * features

    # Update bias
    bias += learnrate * error

    return weights, bias

# Data
features = np.array([[0.1, 0.5, 0.2], [0.2, 0.3, 0.1], [0.7, 0.4, 0.2], [0.1, 0.4, 0.3]])
targets = np.array([0, 1, 0, 1])
epochs = 10
learnrate = 0.1
errors = []
weights = np.array([0.4, 0.2, 0.6])
bias = 0.5

for e in range(epochs):
    for x, y in zip(features, targets):
        pred = predict(x, weights, bias)
        weights, bias = gradient_descent(x, y, weights, bias, learnrate, pred)

    # Printing out the log-loss error on the training set
    out = predict(features, weights, bias)
    loss = np.mean([cross_entropy(y, o) for y, o in zip(targets, out)])
    errors.append(loss)
    print("\n========== Epoch", e, "==========")
    print("Average loss: ", loss)

# Plotting the gradient descent graph
plt.plot(errors)
plt.xlabel('Epochs')
plt.ylabel('Cross-Entropy Loss')
plt.title('Gradient Descent: Cross-Entropy Loss Over Epochs')
plt.show()
```

```python
import numpy as np

# Activation function: Sigmoid
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of the sigmoid function
def sigmoid_derivative(x):
    return x * (1 - x)

# Loss function: Mean Squared Error
def mse_loss(y_true, y_pred):
    return ((y_true - y_pred) ** 2).mean()

# Training data: AND logic gate
inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
targets = np.array([[0], [0], [0], [1]])

# Initialize weights and bias
weights = np.random.rand(2, 1)
bias = np.random.rand(1)

# Hyperparameters
epochs = 10000
learning_rate = 0.1

# Training the neural network
for epoch in range(epochs):
    # Forward pass
    z = np.dot(inputs, weights) + bias
    predictions = sigmoid(z)

    # Compute loss
    loss = mse_loss(targets, predictions)

    # Backward pass
    error = targets - predictions
    d_weights = np.dot(inputs.T, error * sigmoid_derivative(predictions))
    d_bias = np.sum(error * sigmoid_derivative(predictions), axis=0)

    # Update weights and bias
    weights += learning_rate * d_weights
    bias += learning_rate * d_bias

    if epoch % 1000 == 0:
        print(f"Epoch {epoch}, Loss: {loss}")

# Testing the neural network
print("Testing the neural network:")
for input in inputs:
    z = np.dot(input, weights) + bias
    prediction = sigmoid(z)
    print(f"Input: {input}, Prediction: {prediction}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid activation function and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Loss function: Binary Cross-Entropy
def binary_cross_entropy(y_true, y_pred):
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

# Generate synthetic dataset
np.random.seed(0)
num_samples = 100
donkey_features = np.random.rand(num_samples//2, 3) * [0.5, 0.7, 0.8] + [0.3, 0.2, 0.6]
other_features = np.random.rand(num_samples//2, 3) * [0.4, 0.5, 0.6] + [0.1, 0.3, 0.2]
X = np.vstack((donkey_features, other_features))
y = np.hstack((np.ones(num_samples//2), np.zeros(num_samples//2))).reshape(-1, 1)

# Shuffle the dataset
indices = np.arange(num_samples)
np.random.shuffle(indices)
X = X[indices]
y = y[indices]

# Network parameters
input_size = 3
hidden_size = 4
output_size = 1
epochs = 10000
learning_rate = 0.1

# Initialize weights and biases
weights_input_hidden = np.random.rand(input_size, hidden_size)
bias_hidden = np.random.rand(hidden_size)
weights_hidden_output = np.random.rand(hidden_size, output_size)
bias_output = np.random.rand(output_size)

# Training the neural network
losses = []

for epoch in range(epochs):
    # Forward pass
    hidden_input = np.dot(X, weights_input_hidden) + bias_hidden
    hidden_output = sigmoid(hidden_input)
    final_input = np.dot(hidden_output, weights_hidden_output) + bias_output
    final_output = sigmoid(final_input)

    # Compute loss
    loss = binary_cross_entropy(y, final_output)
    losses.append(loss)

    # Backward pass
    error_output = y - final_output
    delta_output = error_output * sigmoid_derivative(final_output)

    error_hidden = np.dot(delta_output, weights_hidden_output.T)
    delta_hidden = error_hidden * sigmoid_derivative(hidden_output)

    # Update weights and biases
    weights_hidden_output += learning_rate * np.dot(hidden_output.T, delta_output)
    bias_output += learning_rate * np.sum(delta_output, axis=0)
    weights_input_hidden += learning_rate * np.dot(X.T, delta_hidden)
    bias_hidden += learning_rate * np.sum(delta_hidden, axis=0)

    if epoch % 1000 == 0:
        print(f"Epoch {epoch}, Loss: {loss}")

# Plotting the loss over epochs
plt.plot(losses)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Loss Over Epochs')
plt.show()

# Testing the neural network
def predict(x):
    hidden_input = np.dot(x, weights_input_hidden) + bias_hidden
    hidden_output = sigmoid(hidden_input)
    final_input = np.dot(hidden_output, weights_hidden_output) + bias_output
    final_output = sigmoid(final_input)
    return final_output > 0.5  # Return class 1 if the output is greater than 0.5

# Example test cases
test_cases = np.array([[0.45, 0.25, 0.75],  # Likely a donkey
                       [0.35, 0.55, 0.35],  # Likely not a donkey
                       [0.50, 0.20, 0.80],  # Likely a donkey
                       [0.20, 0.45, 0.25]]) # Likely not a donkey

predictions = predict(test_cases)

# Displaying test results
for i, test_case in enumerate(test_cases):
    print(f"Test Case {i + 1}: {test_case}, Prediction: {'Donkey' if predictions[i] else 'Not Donkey'}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the network architecture
input_size = 3
hidden_size = 4
output_size = 1

# Function to draw the neural network
def draw_neural_network(ax, input_size, hidden_size, output_size):
    # Define node positions
    input_layer_x = np.full(input_size, 0.1)
    input_layer_y = np.linspace(0, 1, input_size)

    hidden_layer_x = np.full(hidden_size, 0.5)
    hidden_layer_y = np.linspace(0, 1, hidden_size)

    output_layer_x = np.full(output_size, 0.9)
    output_layer_y = np.linspace(0, 1, output_size)

    # Draw nodes
    ax.scatter(input_layer_x, input_layer_y, s=100, label='Input Layer')
    ax.scatter(hidden_layer_x, hidden_layer_y, s=100, label='Hidden Layer')
    ax.scatter(output_layer_x, output_layer_y, s=100, label='Output Layer')

    # Draw edges
    for i in range(input_size):
        for j in range(hidden_size):
            ax.plot([input_layer_x[i], hidden_layer_x[j]], [input_layer_y[i], hidden_layer_y[j]], 'k-', lw=0.5)

    for j in range(hidden_size):
        for k in range(output_size):
            ax.plot([hidden_layer_x[j], output_layer_x[k]], [hidden_layer_y[j], output_layer_y[k]], 'k-', lw=0.5)

    # Set the labels and title
    ax.set_xticks([])
    ax.set_yticks([])
    ax.legend()
    ax.set_title('Neural Network Architecture')

# Create the plot
fig, ax = plt.subplots(figsize=(8, 6))
draw_neural_network(ax, input_size, hidden_size, output_size)
plt.show()
```
